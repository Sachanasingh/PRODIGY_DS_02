```python
plt.figure(figsize=(7,5))
sns.boxplot(data=df['Viewership'])
plt.grid()
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
sns.boxplot(y=df['Revenue (USD)'], ax=axes[0], color="blue").set(title="Revenue Outliers")
sns.boxplot(y=df['Average Player Salary (USD)'], ax=axes[1], color="orange").set(title="Average Salary Outliers")
sns.boxplot(y=df['Viewership'], ax=axes[2], color="green").set(title="Viewership Outliers")
plt.tight_layout()
plt.show()
```

```python
# Exploring Relationships
# Revenue vs Average Salary
sns.scatterplot(data=df, x='Average Player Salary (USD)', y='Revenue (USD)', hue='Sport', palette="viridis")
plt.title("Revenue vs Average Salary")
plt.xlabel("Average Player Salary (USD)")
plt.ylabel("Revenue (USD)")
plt.legend(title="Sport")
plt.show()
```

```python
# Revenue vs Viewership
sns.scatterplot(data=df, x='Viewership', y='Revenue (USD)', hue='Sport', palette="coolwarm")
plt.title("Revenue vs Viewership")
plt.xlabel("Viewership (Millions)")
plt.ylabel("Revenue (USD)")
plt.legend(title="Sport")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='Viewership', y='Sport', hue='Sport', palette="coolwarm")
plt.title("Viewership vs Sport")
plt.xlabel("Viewership (Millions)")
plt.ylabel("Sport")
plt.legend(title="Sport", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
top_countries = df.groupby('Country')['Revenue (USD)'].sum().sort_values(ascending=False).head(10)
top_countries.plot(kind='bar', color="pink")
plt.title("Top 10 Countries by Revenue")
plt.xlabel("Country")
plt.ylabel("Revenue (USD)")
plt.xticks(rotation=45)
plt.show()
```
